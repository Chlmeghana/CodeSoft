```python
#Step 4: Train The Data
x=d.drop('Sales',axis=1)
y=d["Sales"]
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25) # Storing the trained and Tested Datasets
```

```python
print(x_train,x_test,y_train,y_test)
```

```python
print(d)
```

```python
#Step 5:Perform Linear Regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
prediction=lr.predict(x_test)
rmse=np.sqrt(mean_squared_error(y_test,prediction))
rmse#Since the output is 1.7, the prediction is 99% accurate
```
